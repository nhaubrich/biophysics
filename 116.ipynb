```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def saddle(N,z):
    return np.sqrt(N/(2*np.pi*(1-z**2))) * np.exp(N*np.log(1/np.sqrt(1-z**2)) - N*np.arctanh(z)*z)
```

```python
N=1000
samples=100000

u = np.random.randint(2,size=(N,samples))
u = 2*u-1
v = np.random.randint(2,size=(N,samples))
v = 2*v-1

a=plt.hist(np.einsum('ij,ij->j',u,v)/N,density=True,stacked=True,range=(-.2,.2),bins=200+1)
z_val = np.linspace(-.15,.15,num=500)
plt.plot(z_val, saddle(N,z_val),'r-',lw=2)
plt.xlim((-.15,.15))
fig = plt.gcf()
fig.set_size_inches(8, 4)
```
